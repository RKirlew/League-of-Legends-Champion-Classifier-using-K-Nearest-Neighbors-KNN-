```python
import pandas as pd
df = pd.DataFrame(pd.read_excel("Champs.xlsx"))
df.head()
```

```python
print(df['Difficulty'])
print(df['Role'])
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Difficulty'] = le.fit_transform(df['Difficulty'])
print(df['Difficulty'])

le_role = LabelEncoder()
df['Role'] = le_role.fit_transform(df['Role'])
print(df['Role'])

le_range = LabelEncoder()
df['Range type'] = le_range.fit_transform(df['Range type'])
print(df['Range type'])

le_classes = LabelEncoder()
le_classes.fit(df['Classes'])
print(le_classes.classes_)
df['Classes'] = le_classes.transform(df['Classes'])
print(df['Classes'])
```

```python
scaler = StandardScaler()
scaler.fit(df[['Base HP', 'Base mana', 'Difficulty']])
normalized_features = scaler.transform(df[['Base HP', 'Base mana', 'Difficulty']])
normalized_df = pd.DataFrame(normalized_features, columns=['Base HP', 'Base mana', 'Difficulty'])
print(normalized_df)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn import neighbors
from sklearn.model_selection import train_test_split
```

```python
X = pd.concat([normalized_df, df[['Role', 'Range type']]], axis=1)
y_class = df['Classes']
y_role = df['Role']

X_train, X_test, y_train_class, y_test_class = train_test_split(X, y_class, test_size=0.2, random_state=42)

X_train, X_test, y_train_role, y_test_role = train_test_split(X, y_role, test_size=0.2, random_state=42)

print(f'X_train shape: {X_train.shape}, y_train_class shape: {y_train_class.shape}')
print(f'X_test shape: {X_test.shape}, y_test_class shape: {y_test_class.shape}')
```

```python
neigh = KNeighborsClassifier(n_neighbors=4)
neigh.fit(X_train, y_train_class)
```

```python
y_pred = neigh.predict(X_test)

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test_class, y_pred)
print(f'Accuracy: {accuracy}')
```

```python
def predict_champion_class(input_features, knn_model):
    """
    Predicts the champion's class based on input features.

    Args:
    input_features (dict): A dictionary with keys 'Base HP', 'Base mana', 'Difficulty', 'Role', 'Range type'.
    knn_model (KNeighborsClassifier): A trained KNN classifier.

    Returns:
    str: Predicted champion class.
    """

    # Encoding categorical features
    le_difficulty = LabelEncoder()
    le_role = LabelEncoder()
    le_range = LabelEncoder()

    difficulty_encoded = le_difficulty.fit_transform([input_features['Difficulty']])[0]
    role_encoded = le_role.fit_transform([input_features['Role']])[0]
    range_encoded = le_range.fit_transform([input_features['Range type']])[0]

    # Convert input to DataFrame
    input_df = pd.DataFrame([{
        'Base HP': input_features['Base HP'],
        'Base mana': input_features['Base mana'],
        'Difficulty': difficulty_encoded,
        'Role': role_encoded,
        'Range type': range_encoded
    }])

    # Assuming `scaler` is already trained
    scaler = StandardScaler()
    input_df[['Base HP', 'Base mana']] = scaler.fit_transform(input_df[['Base HP', 'Base mana']])

    # Predict class
    predicted_class = knn_model.predict(input_df)
    return le_difficulty.inverse_transform([predicted_class])[0]
```
